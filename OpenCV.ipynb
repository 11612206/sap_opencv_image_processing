```python
def show(image):
    cv2.imshow("temp",image)
    cv2.waitKey(0)
```

```python
# FILTER MANIPULATIONS, SMOOTHING

ret,thresh=cv2.threshold(img2,200,255,cv2.THRESH_BINARY_INV)
show(thresh)
plt.clf()
plt.imshow(thresh,'gray')
```

```python
blur=cv2.blur(thresh,(3,3))
box=cv2.boxFilter(thresh,-1,(3,3),normalize=False)
gaussian=cv2.GaussianBlur(thresh,(3,3),1)
median=cv2.medianBlur(thresh,5)
```

```python
result=np.hstack((thresh,blur,box,gaussian,median))
show(result)
plt.clf()
plt.imshow(result,'gray')
```

```python
# ERODE/DILATE, OPEN/CLOSE, GRADIENT,HAT

kernel=np.ones((3,3),np.uint8)
erosion=cv2.erode(thresh,kernel,iterations=1)
#show(erosion)
dil=cv2.dilate(thresh,kernel,iterations=1)

opening=cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel)
closing=cv2.morphologyEx(thresh,cv2.MORPH_CLOSE,kernel)
```

```python
result2=np.hstack((thresh,opening,closing))
show(result2)
plt.clf()
plt.imshow(result2,'gray')
```

```python
# GRADIENT: DILATE-EROSION

gradient=cv2.morphologyEx(thresh,cv2.MORPH_GRADIENT,kernel)
show(gradient)
```

```python
that=cv2.morphologyEx(thresh,cv2.MORPH_TOPHAT,kernel)
bhat=cv2.morphologyEx(thresh,cv2.MORPH_BLACKHAT,kernel)
result3=np.hstack((thresh,gradient,that,bhat))
show(result3)
plt.clf()
plt.imshow(result3,'gray')
```
